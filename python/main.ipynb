# Solver

```python
from rubiksCubeSolver import RubiksCubeSolver
rubiksCubeSolver = RubiksCubeSolver()
```

```python
state = rubiksCubeSolver.random_state(seed=0)
sol = rubiksCubeSolver.solve(state)
print(sol)
```

```python
solving_process = rubiksCubeSolver.apply_operations(state, sol)
for state in solving_process:
    print(state)
    print('-' * 20 + '\n')
```
